```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\91630_ij1v5fp\Desktop\py_diwali_sales\Python_Diwali_Sales_Analysis-main\Diwali Sales Data.csv",encoding = "unicode_escape")
df.head()
```

```python
df.drop(["Status","unnamed1"],axis = 1,inplace=True)
```

```python
df.head()
```

```python
#to know number of males and females,so plotting a bar chat for numbers and its count.
ax = sns.countplot(x="Gender",data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df_sales = df.groupby(["Gender"],as_index=False)["Amount"].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x="Gender",y="Amount",data=df_sales)
```

```python
#from the above graphs we can say that , most of the buyers are females and their purchase history is also more!.
```

```python
#now lets check with the Age group
```

```python
#here i got to know buyers according to their age group with gender filter.
ax = sns.countplot(x="Age Group",data=df,hue="Gender")

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#now lets, groupby the age group and visualize it in a barchart
age_sales = df.groupby(["Age Group"],as_index = False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sns.barplot(x="Age Group",y="Amount",data=age_sales)
```

```python
#now from the above table, i've come to know the purchasing power of people with different age groups.
```

```python
#now lets see the state wise purchasing power
```

```python
no_states = sns.countplot(x="State",data=df)
for bars in no_states.containers:
    no_states.bar_label(bars)
    sns.set(rc={"figure.figsize" : (15,6)})
```

```python
state_am = df.groupby(["State"],as_index=False)["Amount"].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='State',y='Amount',data=state_am)
sns.set(rc={"figure.figsize":(15,6)})
```

```python
#from this we can say that purchasing power of uttarpradesh,maharashtra and karnataka are high
```

```python
#now lets check on the marital status and lets see who purchases more
```

```python
ms = sns.countplot(x="Marital_Status",data=df)
for bars in ms.containers:
    ms.bar_label(bars)
```

```python
marital = df.groupby(["Marital_Status", "Gender"], as_index=False)["Amount"].sum().sort_values(by="Amount", ascending=False)

sns.barplot(x="Marital_Status", y="Amount", data=marital, hue="Gender")
sns.set(rc={"figure.figsize": (6, 5)})
```

```python
#From above graphs we can see that most of the buyers are married (women) and they have high purchasing power
```

```python
#now lets look at the occupation
```

```python
oc = sns.countplot(x="Occupation",data=df)
for bars in oc.containers:
    oc.bar_label(bars)
    
sns.set(rc={"figure.figsize" : (15,6)})
```

```python
ocg = df.groupby(["Occupation"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sns.barplot(x="Occupation",y="Amount",data=ocg)
```

```python
#*From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector*
```

```python
pd = sns.countplot(x="Product_Category",data=df)
for bars in pd.containers:
    pd.bar_label(bars)
    
sns.set(rc={"figure.figsize": (20, 7)})
```

```python
ocp = df.groupby(["Product_Category"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sns.barplot(x="Product_Category",y="Amount",data = ocp)
sns.set(rc={"figure.figsize": (12, 8)})
```

```python
#From above graphs we can see that most of the sold products are Food, Clothing and Electronics category
```

```python
pi = df.groupby(["Product_ID"],as_index=False)["Orders"].sum().sort_values(by="Orders",ascending=False).head(10)
sns.barplot(y="Orders",x="Product_ID",data=pi)
```

```python
#Married women age group 26-35 yrs from UP, Maharastra and Karnataka working in IT, Healthcare and Aviation are more likely to buy products from Food, Clothing and Electronics category
```
